```python
def top_tracks(artist_id, auth_info):
    try:
        end_point = f'https://api.spotify.com/v1/artists/{artist_id}/top-tracks'
        response = requests.get(end_point, headers=auth_info, verify=False, timeout=30)
        track_list = response.json()['tracks']
        track_names = [track['name'] for track in track_list]
        track_ids = [track['id'] for track in track_list]
        #Nuevo codigo
        top_tracks_info = [{'name': name, 'id': id} for name, id in zip(track_names, track_ids)]
        # Asegúrate de devolver ambos resultados si lo requiere el flujo de tu código
        return track_names, track_ids
    
    except Exception as e:
        print(f"Error obteniendo top tracks para el artista {artist_id}: {e}")
        return [], []  # Devolver listas vacías en caso de error
        
```

```python
def track_list_generator(artist_id, auth_info):
    try: 
        end_point = f'https://api.spotify.com/v1/artists/{artist_id}/related-artists'
        response = requests.get(end_point, headers=auth_info, verify=False, timeout=30)
        artists_list = response.json()['artists']
        artists_names = [artist['name'] for artist in artists_list]
        artists_ids = [artist['id'] for artist in artists_list]
        full_track_list = [top_tracks(artist, auth_info) for artist in artists_ids]
        return pd.concat(full_track_list).reset_index(drop=True)
    except:
        return response.status_code
```
